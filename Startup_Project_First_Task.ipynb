```python
#Checking for the categorical attributes 
cat_col = []
for x in copy_data.dtypes.index:
    if copy_data.dtypes[x] == 'object':
        cat_col.append(x)
cat_col
```

```python
#Removing the 'Unnamed: 0.1' column
copy_data.drop(['Unnamed: 0.1'],axis=1, inplace=True)
```

```python
#Displaying the first five values of the data
copy_data.head()
```

```python
#List of the columns to be removed
#'closed at' for missing excessive values
#'ROI' for missing exessive values
#'milestones': Not important for analysis
#'created by': Not important for analysis
data_cleaned = copy_data.drop(['closed_at','ROI','milestones','created_by'],axis=1)
```

```python
#Displaying the data after removing the columns
data_cleaned
```

```python
#List of the columns to impute 
columns_to_impute = [
    'homepage_url',
    'twitter_username',
    'logo_url',
    'logo_width',
    'logo_height',
    'first_funding_at',
    'last_funding_at',
    'funding_rounds',
    'funding_total_usd',
]
```

```python
#Imputing the numeric values with median and an empty string for text columns
for column in columns_to_impute:
    if data_cleaned[column].dtype == 'object':
        # For text columns, fill missing values with an empty string
        data_cleaned[column].fillna('', inplace=True)
    else:
        # For numeric columns, fill missing values with the median
        median_value = data_cleaned[column].median()
        data_cleaned[column].fillna(median_value, inplace=True)
```

```python
#Displaying the data after imputing the values
data_cleaned
```

```python
#Checking the missing values of the specified columns
data_cleaned.isnull().sum()
```

```python
#List of columns for encoding
columns_to_one_hot_encode = ['entity_type', 'category_code', 'country_code', 'state_code', 'city', 'region']

# Perform one-hot encoding for selected columns
data_cleaned_encoded = pd.get_dummies(data_cleaned, columns=columns_to_one_hot_encode, drop_first=True)

# Displaying the encoded DataFrame
data_cleaned_encoded
```

```python
#Importing the module
from sklearn.preprocessing import LabelEncoder

# List of columns to label encode (ordinal data)
columns_to_label_encode = ['status']

# Perform label encoding for selected columns
label_encoder = LabelEncoder()
for column in columns_to_label_encode:
    data_cleaned_encoded[column] = label_encoder.fit_transform(data_cleaned_encoded[column])

# Displaying the DataFrame with label encoding applied
data_cleaned_encoded
```
